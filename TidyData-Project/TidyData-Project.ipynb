### Import Libraries

```python
import pandas as pd
import matplotlib as plt
```

### Load Data

```python
#load in Fed Spending data 
fed_data = pd.read_csv('./fed_rd_year&gdp.csv')

display(fed_data)
fed_data.columns
```

### Data Cleaning and Tidy Process
Tidy data by changing column names to numeric year values then melt the data to make it long instead of wide.

```python
# rename cols

#create new empty name dictionary
new_names={}

for i in fed_data.columns:      #loop through all columns
    if '_gdp' in i:             #if it is a year gdp column
        year = i.split('_')[0]  #grab just the year
        new_names[i] = year     #assign it to the new_names dictionary

fed_data_clean = fed_data.rename(columns=new_names) #replace column names with new names
fed_data_clean
```

```python
# melt data
fed_data_clean = pd.melt(fed_data_clean, id_vars=['department'],
                    var_name='Year', value_name='Spending')

#make year numeric
fed_data_clean['Year'] = pd.to_numeric(fed_data_clean['Year'])

fed_data_clean
# Each variable is its own column (department, year, spending)
# Each observation(department) has their own row
# Each observational unit forms its own table
```

### Visualization 1 - Time Series Line Chart
Created a line chart that shows the spending time series broken down by department.

```python
#loop through the departments
for dept in fed_data_clean['department'].unique():
    #filter data to department selected
    dept_data = fed_data_clean[fed_data_clean['department'] == dept]
    #plot for the department 
    plt.plot(dept_data['Year'], dept_data['Spending'], label=dept)

#labels
plt.title('Spending by Department')
plt.xlabel('Year')
plt.ylabel('Spending')
plt.legend(fontsize='small')
plt.show()
```

### Visualization 2 - Bar Chart
Created a bar chart that allows the user to select a year then shows the proportion of spending broken down by department for that particular year.

```python
#select year for the chart
year = 2010
year_fed_data = fed_data_clean[fed_data_clean['Year']==year]

#create share variable
year_fed_data['Share'] = year_fed_data['Spending']/sum(year_fed_data['Spending'])

#plot and labels
plt.barh(year_fed_data['department'], year_fed_data['Share'])
plt.title('Share of Spending by Department')
plt.xlabel('Share of Spending')
plt.ylabel('Department')
plt.show()
```

### Pivot-Table
Created a pivot table that aggregates the median spending by department for each year in the dataset.

```python
#create median spending pivot table
fed_data_pivot = pd.pivot_table(fed_data_clean, values='Spending', index='Year',  aggfunc='median')
fed_data_pivot
```
